```python
import pandas as pd
import seaborn as sns
# import numpy as np
```

```python
data__headers = ["user_id","item_id","rating","timestamp"]
item__headers = ["item_id","item_title"]

data = pd.read_csv("./dataset/ml-100k/u.data" , names=data__headers , sep="\t").drop(columns="timestamp")
item = pd.read_csv("./dataset/ml-100k/u.item" , names=item__headers, sep="|" , encoding = 'latin-1', usecols=[0,1])

# print(data.head())
# print()
# print(item.head())

merged = pd.merge(data,item,on="item_id")
pivot = merged.pivot_table(index="user_id" ,columns="item_title" ,values="rating")
pivot
```

```python
avg_ratings = data.groupby(['item_id']).mean()
avg_ratings = pd.merge(avg_ratings,item,on="item_id")
avg_ratings = avg_ratings.drop(columns=["item_id","user_id"]).sort_values('rating')
# print(avg_ratings.head())

total_ratings = data.groupby(['item_id']).count()
total_ratings = pd.merge(total_ratings,item,on="item_id")
total_ratings = total_ratings.drop(columns=["item_id","user_id"]).sort_values('rating')
total_ratings = total_ratings.rename(columns = {"rating":"number_of_ratings"})
# print(total_ratings.head())

ratings = pd.merge(avg_ratings,total_ratings,on="item_title")
print(ratings.head())

sns.jointplot(data=ratings, x="rating", y="number_of_ratings", kind="reg")
```

```python
def predict(item_name):
    
    item_ratings = pivot[item_name]
    similarity_index = pivot.corrwith(item_ratings)
    similarity_index = pd.DataFrame(similarity_index ,columns=["correlation"])
#     print(similarity_index)
    similarity_index = pd.merge(similarity_index, total_ratings, on="item_title").dropna(axis=0,how="any")
#     print(similarity_index)
    predictions = similarity_index[similarity_index["number_of_ratings"]>100].sort_values("correlation",ascending=False)

    return predictions
```

```python
prediction = predict("101 Dalmatians (1996)")
```
